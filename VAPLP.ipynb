```python
# Import the necessary libraries
# Numbers
import numpy as np                                      # linear algebra
import pandas as pd                                     # Data processing, CSV file I/O (e.g. pd.read_csv)

# Plots
import seaborn as sns                                   # importing seaborn for statistical plots
import matplotlib.pyplot as plt                         # matplotlib.pyplot plots data

# Train test split
from sklearn.model_selection import train_test_split   # Sklearn package's randomized data splitting function

# Scaling related
from sklearn import preprocessing                      # Provides several common utility functions: standardization of the data set
from scipy.stats import zscore                         # Computes standard score relative to the sample mean and standard deviation

# Models
from sklearn.linear_model import LinearRegression      # To model the ordinary least squares Linear Regression
from sklearn.linear_model import Ridge                 # To model Linear least squares with l2 regularization
from sklearn.linear_model import Lasso                 # Linear Model trained with L1 prior as regularizer
from sklearn.preprocessing import PolynomialFeatures   # Generates a new feature matrix consisting of all polynomial combinations of the features
from sklearn.linear_model import ElasticNet            # To model Elastic Net Regression

# Metrics
from sklearn import metrics                            # sklearn metrics module implements utility functions to measure classification performance
from sklearn.metrics import mean_squared_error         # Determines mean squared error regression loss
from sklearn.metrics import mean_absolute_error        # Determines mean absolute errorregressionloss
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=1)

#Checking split of data
print("X training data size: {}".format(X_train.shape))
print("y training data size: {}".format(y_train.shape))
print("X testing data size: {}".format(X_test.shape))
print("y testing data size: {}".format(y_test.shape))
print("Data in dataset".format((len(X_train)/len(df.index))*100))
```

Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)
print("Linear Regression model coefficients:",(lr.coef_))
print("LR Mean Square Error :",mean_squared_error(y_test,y_pred_lr))
print("Root Mean Square Error :",np.sqrt(mean_squared_error(y_test,y_pred_lr)))
print("Mean absolut error:",mean_absolute_error(y_test,y_pred_lr))
```

Ridge Regression

```python
ridge=Ridge(alpha=.3)
ridge.fit(X_train,y_train)
y_pred_ridge=ridge.predict(X_test)

print('Ridge Regression model coefficients:',(ridge.coef_))
print('RR Mean Square Error  :', mean_squared_error(y_test,y_pred_ridge))
print('RR Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,y_pred_ridge)))
print('RR Mean Absolute Error:',mean_absolute_error(y_test,y_pred_ridge))
```

Lasso Regression

```python
lasso=Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
y_pred_lasso=lasso.predict(X_test)

print('Lasso Regression model coefficients:',(lasso.coef_))
print('LR Mean Square Error  :', mean_squared_error(y_test,y_pred_lasso))
print('LR Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,y_pred_lasso)))
print('LR Mean Absolute Error:',mean_absolute_error(y_test,y_pred_lasso))
```

Elastic Regression

```python
en=ElasticNet(alpha=1.0,l1_ratio=0.5)
en.fit(X_train,y_train)
y_pred_en=en.predict(X_test)

print('Elastic Net Regression model coefficients:',(en.coef_))
print('ENR Mean Square Error  :', mean_squared_error(y_test,y_pred_en))
print('ENR Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,y_pred_en)))
print('ENR Mean Absolute Error:',mean_absolute_error(y_test,y_pred_en))
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True)
X_poly=poly.fit_transform(X_scaled)
X_poly.shape
```

```python
lr=LinearRegression()
lr.fit(X_train_poly,y_train)
y_pred_lr=lr.predict(X_test_poly)
```

```python
print('Polynomial Regression model coefficients:',(lr.coef_))
print('PR Mean Square Error  :', mean_squared_error(y_test,y_pred_lr))
print('PR Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,y_pred_lr)))
print('PR Mean Absolute Error:',mean_absolute_error(y_test,y_pred_lr))
```

Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
#dt = DecisionTreeRegressor(random_state = 0)
dt=DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=None, min_samples_split=4, min_samples_leaf=5, min_weight_fraction_leaf=0.0, max_features=None, random_state=1, max_leaf_nodes=10)
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)

#print('Elastic Net Regression model coefficients:',(dt.coef_))
print('ENR Mean Square Error  :', mean_squared_error(y_test,y_pred_dt))
print('ENR Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,y_pred_dt)))
print('ENR Mean Absolute Error:',mean_absolute_error(y_test,y_pred_dt))
```

Random Forest Regressor
